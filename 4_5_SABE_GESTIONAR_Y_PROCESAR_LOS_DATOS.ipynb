Antes de resolver los puntos que se nos presentan, necesitamos importar la librería pandas y guardar la base de datos en una variable.

```python
import pandas as pd
data_base = pd.read_csv('insurance.csv')
data_base
```

Una vez realizado este paso, ya podemos empezar a resolver las cuestiones. La primera nos pide aplicar la función Describe (), sumar los bmi y calcular la media de los charges para el intervalo de filas desde la 45 hasta la 68. Bueno, empezamos por crear una nueva base de datos que esté formada sólo por este intervalo de filas:

```python
data_base_2 = data_base.iloc [45:69,]
data_base_2
```

Nota: recuerda que el 69 no se incluye ya que estamos indexando una lista
Aplicamos, ahora, la función Describe () y obtenemos la siguiente tabla:

```python
data_base_2['bmi'].sum()
```

El resultado que obtenemos es 756.095.
A continuación, calculamos la media de charges mediante la función mean (). Lo hacemos de esta manera:

Seguimos con el mismo intervalo de filas para el segundo punto. Ahora, nos piden que mostramos los 12 primeros registros de manera ordenada en orden descendente, según los valores de bmi. Bueno, necesitamos aplicar la función sort_values () y la función head (). Lo hacemos así:

En el tercer punto, tenemos que volver a utilizar la base de datos principal data_base. Con el fin de saber cuántas personas fumadoras hay en la zona northwest, habrá agrupar la base de datos en dos niveles: primero, según región y, después, según smoker. El código que necesitamos es:

Ahora, necesitaremos crear otra base de datos a partir de la principal data_base con filas que sólo tengan valores de bmi superiores a 35 y que las personas no sean fumadoras. El código que necesitamos es:

```python
data_base_3 = data_base[(data_base['bmi']>35) & (data_base['smoker']=='no')] 
data_base_3.head(12)
```

La tabla que resulta, después de ejecutar el código, es esta:

Fijémonos que aquí hemos hecho uso de la indexación a través de la introducción de una serie de condiciones: la primera es decir que los valores de la variable bmi deben ser superiores a 35 y que los valores de la variable smoker sólo pueden tener el valor de no. Todas aquellas observaciones que no cumplan estrictamente las dos condiciones no se cargarán a la nueva tabla que hemos llamado data_base_3. Las filas deben cumplir las dos condiciones. En el caso de que sólo cumplan una no serán seleccionadas. Para que sí sean seleccionadas sólo cumpliendo una de las dos condiciones, tenemos que cambiar el código de esta manera:

```python
data_base_3 = data_base[(data_base['bmi']> 35) | (data_base['smoker'] == 'no')] 
data_base_3.head(12)
```

Lo que hemos hecho es sencillamente cambiar el valor del operador condicional. En el primer caso, teníamos el operador &, que implica que las dos condiciones se deben cumplir. En el segundo caso, en cambio, tenemos el operador |, que implica que sólo cumpliendo una de las dos condiciones, la fila será seleccionada.

Si tenemos que añadir la edad superior a 25 como condición obligatoria, manteniendo que, simplemente el valor bmi sea superior a 35 o que la persona no sea fumadora, tendremos que modificar el código y escribirlo así:

```python
data_base_3 = data_base[((data_base['bmi']> 35) | (data_base['smoker'] == 'no')) & (data_base [ 'age']> 25)]
data_base_3.head (12)
```

Ahora, estamos haciendo uso de los operadores & y | en el mismo código. Con los paréntesis, lo que hacemos es agrupar las condiciones. De hecho, hay dos condiciones que debe cumplir. 

La primera es:

```python
(data_base ['bmi']>35) | (data_base['smoker'] == 'no')
```

Y la segunda es: 

```python
data_base['age']> 25
```

Ambas deben cumplirse, pero, para que se cumpla la primera, simplemente una de las dos opciones que contiene sea cierta.
